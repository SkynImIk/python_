```python
import pandas as pd
import matplotlib.pyplot as plt

equipment_df = pd.read_csv('russia_losses_equipment.csv')

equipment_df.drop_duplicates()
equipment_df=equipment_df.fillna(0)

display(equipment_df.describe())

group_equipment = equipment_df.groupby('helicopter')['APC'].mean()
display(group_equipment)

display(equipment_df.sort_values('APC', ascending=True))

plt.figure(figsize=(15, 5))
plt.plot(equipment_df['day'], equipment_df['APC'], 'r')
plt.xlabel('day')
plt.ylabel('APC')
plt.title('Динаміка втрат техніки')
# plt.legend()
plt.show()
```
